```python
#查看原始图片
img = cv2.imread(r'D:\git\Untitled Folder\2017-07-16-18-29-14000000.jpg',1)
plt.imshow(img)
```

```python
#单张图片运行结果的可视化
lable = datas[0]['labels']
gt_boxes = lable['gt_boxes']
gt_classes = lable['gt_classes']
gt_ignore_boxes = lable['gt_ignore_boxes']
gt_ignore_classes = lable['gt_ignore_boxes']

text = 'car'

#需要判断坐标阈值，是否落到图片外
for gt_box in gt_boxes:
    xmin = int(gt_box[0])
    ymin = int(gt_box[1])
    xmax = int(gt_box[2])
    ymax = int(gt_box[3])
    print(xmin,ymin,xmax,ymax)
    cv2.rectangle(img, (xmin,ymin), (xmax,ymax), (0,255,0), 4)
    font = cv2.FONT_HERSHEY_COMPLEX_SMALL
    cv2.putText(img, text, (xmin, ymax-6), font, 1, (0,0,255), 1)
```
